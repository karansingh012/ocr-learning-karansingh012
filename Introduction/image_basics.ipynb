```python
# Black & White
# Grayscale conversion utility
# color ke liye 3 channel hota hai (R, G, B) (0-255)*3
# Grayscale me sirf 1 channel hota hai (intensity) (0-255)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Image read
img = cv2.imread('fruit.png')
cv2.imshow('Original Image', img)
cv2.waitKey(0)
```

```python
# Grayscale conversion
import cv2
img = cv2.imread('fruit.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray_img.shape)  # (height, width)
cv2.imshow('Grayscale Image', gray_img)
cv2.waitKey(0)
```

```python
import cv2
img = cv2.imread('fruit.png')
img_resized = cv2.resize(img, (300, 200))
cv2.imshow('Resized Image', img_resized)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np 
img = cv2.imread('fruit.png')
img_resized = cv2.resize(img,(img.shape[1]//2, img.shape[0]//2)) # Resize by half
cv2.imshow('Resized Image', img_resized)
print(img_resized.shape)
cv2.waitKey(0)
```

```python
# flip image
# three codes for flip
# 0 - vertical flip
# 1 - horizontal flip
# -1 - both axes flip
import cv2
img = cv2.imread('fruit.png')
flipped_img = cv2.flip(img, 0)  # 1 for horizontal flip
cv2.imshow('Flipped Image', flipped_img)
cv2.waitKey(0)

```

```python
# crop image
import cv2
img = cv2.imread('fruit.png')
cropped_img = img[50:200, 100:300]  # Crop region ## basic slicing X width y height 
cv2.imshow('Cropped Image', cropped_img)
cv2.waitKey(0)
```

```python
## how to save image
import cv2
img = cv2.imread('fruit.png')
cv2.imwrite('saved_fruit.png', img)
cv2.imshow('Saved Image', img)
cv2.waitKey(0)
```

```python
# Drawing Shapes and Text on Images
import cv2
import numpy as np
# read an image 
img = np.zeros((500, 500, 3))
## RECTANGLE
cv2.rectangle(img,pt1=(50, 50), pt2=(200, 200), color=(0, 255, 0), thickness=3)  # Green rectangle with thickness 3
## circle
cv2.circle(img, center=(300, 300), radius=50, color=(255, 0, 0), thickness=-1)  # Filled blue circle
## line
cv2.line(img, pt1=(0, 0), pt2=(500, 500), color=(0, 0, 255), thickness=2)  # Red line
## text
cv2.putText(img, text='Hello OpenCV', org=(50, 400), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(255, 255, 255), thickness=2)
## show image
cv2.imshow('Black Image', img)
cv2.waitKey(0)
```
