```python
import cv2
import numpy as np

img = cv2.imread("/Users/karansingh/ocr-learning/Airplane.jpg")
flag = False 
ix = -1
iy = -1
def crop(event,x,y,flags,param):
    global ix,iy,flag,img

    if event == 1:
        flag = True 
        ix = x 
        iy = y

    # elif event == 0:
    #     if flag == True:
    #         cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,0,0),thickness=1)
    elif event == 4:
        fx = x
        fy = y
        flag = False
        cv2.rectangle(img,pt1=(ix,iy),pt2=(x,y),color=(0,0,0),thickness=5)
        ## Crop tool 
        cropped = img[iy:fy,ix:fx]
        cv2.imshow("Cropped",cropped)
        cv2.waitKey(0)

cv2.namedWindow('Image')
cv2.setMouseCallback('Image',crop)
while True:

    cv2.imshow("Image", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()


```
