```python
#Live Direct drawing on images using OpenCV
import cv2
import numpy as np
img = np.zeros((512,512,3), np.uint8)


while True:

    cv2.imshow('Image', img)
    if cv2.waitKey(1) & 0xFF == ord('q'): # for immediate break
        break 

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def draw_circle(event, x, y, flags, param):
    # print("Event triggered:")
    # print(event)
    if event == 0: # left mouse button down event
        cv2.circle(img, center=(x,y), radius= 5, color=(0,255,0), thickness=-1)
        print("mouse moved to:", x, y)
    cv2.namedWindow(winname='Image')
    cv2.setMouseCallback('Image', draw_circle)
    # img = np.zeros((512,512,3), np.uint8)
    img = cv2.imread('fruit.png')
    while True:
        cv2.imshow('Image', img)
        if cv2.waitKey(1) & 0xFF == ord('q'): # for immediate break
            break
    cv2.destroyAllWindows()

```
