```python
## Minimal working code 
import cv2
import pytesseract


img = cv2.imread('/Users/karansingh/ocr-learning/pytesseractmain/text_image.jpg')
text = pytesseract.image_to_string(img)
print(text)
```

```python
# üü¶ STEP 4: WHY OPENCV + TESSERACT?
# Raw image ‚ùå
# Preprocessed image ‚úÖ
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
text = pytesseract.image_to_string(gray)
print(text)

# üëâ Accuracy improves immediately.
```

# üü¶ STEP 5: PAGE SEGMENTATION MODES (PSM) üî•

```python
# PSM tells Tesseract how text is arranged.
# Common PSM modes:
# PSM         Use case
# 3          Default (block of text)
# 6          Assume a single uniform block of text
# 7          Treat the image as a single text line
# 11         Sparse text with OSD  
```

# üîπ Use PSM in code

```python
custom_config = r'--psm 6'
text = pytesseract.image_to_string(gray, config=custom_config)
```

# üü¶ STEP 6: OCR ENGINE MODE (OEM)

```python
# OEM decides OCR engine type.

# OEM options:
# 	‚Ä¢	1 ‚Üí LSTM neural net (BEST)
# 	‚Ä¢	3 ‚Üí Default (LSTM + legacy)

# üîπ Use OEM + PSM together

custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(gray, config=custom_config)


# üëâ This is production-style OCR.
```

# # üü¶ STEP 7: LANGUAGE SETTING

```python
# üîπ Specify language
 # Default language = English.
text = pytesseract.image_to_string(gray, lang='eng')
# For Hindi:
text = pytesseract.image_to_string(gray, lang='hin')
# For multiple languages:
text = pytesseract.image_to_string(gray, lang='eng+hin')

# üß† IMPORTANT OCR RULE (remember this)

# 80% OCR accuracy comes from preprocessing, not from the OCR engine.

# That‚Äôs why OpenCV is critical.
```
