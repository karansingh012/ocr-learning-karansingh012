What is an OCR Pipeline?
* A pipeline means text flows through fixed stages:
Input Image
 ‚Üí Preprocessing (OpenCV)
 ‚Üí OCR (Tesseract)
 ‚Üí Raw Text
 ‚Üí Clean Text

# üü¶ FULL OCR PIPELINE (STEP-BY-STEP)
üîπ STEP 1: Read Image
üîπ STEP 2: Convert to Grayscale
üîπ STEP 3: Noise Removal
üîπ STEP 4: Thresholding (VERY IMPORTANT üî•)
üîπ STEP 5: Morphological Operations
üîπ STEP 6: OCR (Tesseract)
*  OPTIONAL (but powerful): Deskew Image
üßπ STEP 7: Text Post-Processing (IMPORTANT)
FINAL PIPELINE (Remember this üî•)
Image
 ‚Üí Grayscale
 ‚Üí Blur
 ‚Üí Threshold
 ‚Üí Morphology
 ‚Üí OCR
 ‚Üí Clean Text


```python
import cv2
import pytesseract

img = cv2.imread("/Users/karansingh/ocr-learning/pytesseract/vlrYO.jpg")
```

```python
# Preprocess the image (grayscale, thresholding, etc.)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
_, thresh = cv2.threshold(
    blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Why?
# 	‚Ä¢	Converts image to black & white
# 	‚Ä¢	Makes text sharp
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
processed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
# Why?
# 	‚Ä¢	Connects broken characters
# 	‚Ä¢	Improves readability
```

```python
custom_config = r'--oem 3 --psm 6'
text = pytesseract.image_to_string(processed, config=custom_config)
print(text)

# üëâ This is where text is extracted.
```

```python
coords = cv2.findNonZero(thresh)
angle = cv2.minAreaRect(coords)[-1]

if angle < -45:
    angle = -(90 + angle)
else:
    angle = -angle

(h, w) = img.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, angle, 1.0)
processed = cv2.warpAffine(img, M, (w, h),
                           flags=cv2.INTER_CUBIC,
                           borderMode=cv2.BORDER_REPLICATE)
```

```python
import re

clean_text = re.sub(r'[^\x00-\x7F]+', ' ', text)
clean_text = re.sub(r'\s+', ' ', clean_text)

print(clean_text)
```
