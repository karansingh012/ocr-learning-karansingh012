```python
## Drawing a line
import cv2
import numpy as np
img = np.zeros((512,512,3), np.uint8) ## create a black image
if img is None:
    print("Image not created")
else:
    print("Image created")
cv2.line(img, pt1 =(0,0), pt2 =(511,511), color=(255,0,0), thickness=5) ## draw a blue line with thickness of 5 px
cv2.imshow('Line Image', img) ## display the image
cv2.waitKey(0) ## wait for a key press
cv2.destroyAllWindows() ## close all windows
```

```python
## Drawing a rectangle
import cv2
import numpy as np
img = np.zeros((512,512,3), np.uint8) ## create a black image
if img is None:
    print("Image not created")
else:
    print("Image created")
cv2.rectangle(img, pt1 =(384,0), pt2 =(510,128), color=(0,255,0), thickness=3) ## draw a green rectangle with thickness of 3 px
cv2.imshow('Rectangle Image', img) ## display the image
cv2.waitKey(0) ## wait for a key press
cv2.destroyAllWindows() ## close all windows 
```

```python
## Drawing a circle
import cv2
import numpy as np
img = np.zeros((512,512,3), np.uint8) ## create a black image
if img is None:
    print("Image not created")
else:
    print("Image created")
cv2.circle(img, center =(447,63), radius= 63, color=(0,0,255), thickness=-1) ## draw a filled red circle
cv2.imshow('Circle Image', img) ## display the image
cv2.waitKey(0) ## wait for a key press
cv2.destroyAllWindows() ## close all windows    
```

```python
# Adding text to an image
import cv2
import numpy as np
img = np.zeros((512,512,3), np.uint8) ## create a black
if img is None:
    print("Image not created")
else:
    print("Image created")
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, text='OpenCV Tutorial', org=(10,500), fontFace=font, fontScale=1, color=(255,255,255), thickness=2, lineType=cv2.LINE_AA)
cv2.imshow('Text Image', img) ## display the image
cv2.waitKey(0) ## wait for a key press
cv2.destroyAllWindows() ## close all windows   
```
