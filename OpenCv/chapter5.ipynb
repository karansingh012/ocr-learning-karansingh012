```python
# Gaussian Blur = Gaussian blur is a technique used in image processing to reduce noise and detail in an image by applying a Gaussian function to the pixel values. It smooths the image while preserving edges, resulting in a softer appearance. Gaussian blur is commonly used in photography, computer vision, and graphics applications to enhance image quality and reduce artifacts.
# why Gaussian Blur = Gaussian blur is used to reduce noise, enhance image quality, and improve the performance of computer vision algorithms by smoothing images while preserving edges.
# how Gaussian Blur works = Gaussian blur works by convolving an image with a Gaussian function, which assigns weights to neighboring pixels based on their distance from the center pixel. The closer a pixel is to the center, the higher its weight, resulting in a weighted average that smooths the image while preserving important features like edges.
# applications of Gaussian Blur = 1. Noise Reduction, 2. Image Smoothing, 3. Preprocessing for Computer Vision, 4. Artistic Effects, 5. Depth of Field Simulation 6. Medical Imaging, 7. Object Detection and Recognition  8. Video Processing 9. Texture Analysis 10. Image Segmentation 

# blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigmaX) 
import cv2
import numpy as np
img = cv2.imread('/Users/karansingh/ocr-learning/OpenCv/fruit1.png')
if img is None:
    print("Image not found")
else:
    blurred_img = cv2.GaussianBlur(img, (7,7), 4)
    cv2.imshow('Original Image', img)
    cv2.imshow('Blurred Image', blurred_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

# Sharping

```python
## cv2.filter2D(src, ddepth, kernel) 
import cv2
import numpy as np
img = cv2.imread('/Users/karansingh/ocr-learning/OpenCv/fruit_median_blurred.png')
if img is None:
    print("Image not found")
else:
    # Define a sharpening kernel
    sharpening_kernel = np.array([[0, -1, 0],
                                  [-1,  9, -1],
                                  [0, -1, 0]])
    sharpened_img = cv2.filter2D(img, -1, sharpening_kernel)
    cv2.imshow('Original Image', img)
    cv2.imshow('Sharpened Image', sharpened_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
