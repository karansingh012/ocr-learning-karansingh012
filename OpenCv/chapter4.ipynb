# Video function using cap.py

```python
import cv2
import numpy as np
# Capture video from the default webcam (0)
cap = cv2.VideoCapture(0) 
while True:
    ret, frame = cap.read()  # Read a frame from the webcam
    if not ret:
        print("Failed to grab frame")
        break  # Exit the loop if there are no more frames

    cv2.imshow('Webcam Feed', frame)  # Display the frame in a window

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'): ## wait for 1 ms 113 is ascii of q 
        print("Quitting...")
        break
```

# Saving video

```python
# cv.videoWriter(filename, fourcc, fps, frameSize)
# fourcc = cv2.VideoWriter_fourcc(*'XVID') # codec
# fps = 20.0
# frameSize = (640, 480)
import cv2

cammera = cv2.VideoCapture(0)
frame_width = int(cammera.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cammera.get(cv2.CAP_PROP_FRAME_HEIGHT))
codec = cv2.VideoWriter_fourcc(*'XVID')
recoreded_video = cv2.VideoWriter('output.avi', codec, 20.0, (frame_width, frame_height))
while True:
    ret, frame = cammera.read()
    if not ret:
        print("Failed to grab frame")
        break
    recoreded_video.write(frame)
    cv2.imshow('Webcam Feed', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Quitting...")
        break
 
```
