# Loading an image 

```python
import cv2
import numpy as np
# 1 = color image
# 0 = grayscale image
# -1 = including alpha channel
## reading image 
img = cv2.imread('/Users/karansingh/ocr-learning/Introduction/fruit.png')
## displaying image
if img is None:
    print("Image not found")
else:
    print("Image found")
```

# Display an image

```python
img = cv2.imread('/Users/karansingh/ocr-learning/Introduction/fruit.png')
if img is not None:
    cv2.imshow('Output Image', img) ## open a window and display image
    cv2.waitKey(0) ## wait for a key press to close the window
    cv2.destroyAllWindows() ## close all windows
else:
    print("Image not found") ## check if image is loaded successfully
```

# Save an image 

```python
import cv2
img = cv2.imread('/Users/karansingh/ocr-learning/Introduction/fruit.png')
if img is not None:
    success = cv2.imwrite('output_fruit.png', img) ## save the image to a file
    if success:
        print("Image saved successfully")
    else:
        print("Image not found, cannot save")
else:
    print("Image not found, cannot save")
```

# Gray scale conversion

```python
img = cv2.imread('/Users/karansingh/ocr-learning/Introduction/fruit.png')
if img is not None:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) ## convert to grayscale
    cv2.imshow('Grayscale Image', gray_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found")
```
