# image transformation and manipulation technique

# Resizing

```python
import cv2
import numpy as np 
#(image path, width, height, color channels)
img = cv2.imread('/Users/karansingh/ocr-learning/Introduction/fruit.png')
if img is not None:
    img_resized = cv2.resize(img,(300,300)) ## resize image to 500x500
    cv2.imshow('Original Image', img)
    cv2.imshow('Resized Image', img_resized)
    cv2.imwrite('/Users/karansingh/ocr-learning/Introduction/fruit_resized.png', img_resized) ## save resized image
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found") ## check if image is loaded successfully
```

# Cropping

```python
## Cropped Image
# cropped_image = img[startY:endY, startX:endX]
import cv2
img = cv2.imread('/Users/karansingh/ocr-learning/OpenCv/fruit1.png')
if img is not None:
    cropped_img = img[50:250, 100:300] ## crop image from (100,50) to (300,250)
    cv2.imshow('Cropped Image', cropped_img)
    cv2.imwrite('/Users/karansingh/ocr-learning/Introduction/fruit_cropped.png', cropped_img) ## save cropped image
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found") ## check if image is loaded successfully
```

# Image rotation and flipping

```python
# M = cv2.getRotationMatrix2D((center_x, center_y), angle, scale)
# rotated_image = cv2.warpAffine(image, M, (width, height))
# Center point = width/2, height/2
#  M formula
import cv2
import numpy as np
img = cv2.imread('/Users/karansingh/ocr-learning/OpenCv/fruit1.png')
if img is None:
    print("Image not found")
else:
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, 90, 1.0) ## rotate image by 45 degrees
    rotated_img = cv2.warpAffine(img, M, (w, h))
    cv2.imshow('Rotated Image', rotated_img)
    cv2.imwrite('/Users/karansingh/ocr-learning/Introduction/fruit_rotated.png', rotated_img) ## save rotated image
    cv2.waitKey(0)
    cv2.destroyAllWindows()    
```

# image flipping

```python
## cv2.flip(image, flipCode)
# flipCode = 0: flip vertically
# flipCode > 0: flip horizontally
# flipCode < 0: flip both vertically and horizontally
import cv2
img = cv2.imread('/Users/karansingh/ocr-learning/OpenCv/fruit1.png')
if img is None:
    print("Image not found")
else:
    flipped_horizontal = cv2.flip(img, 1) ## flip image horizontally
    flipped_vertical = cv2.flip(img, 0) ## flip image vertically
    flipped_both = cv2.flip(img, -1) ## flip image both vertically and horizontally
    #flipped_img = np.hstack((flipped_horizontal, flipped_vertical, flipped_both)) ## stack images horizontally
    cv2.imshow('Original', img)
    cv2.imshow('horizontal', flipped_horizontal)
    cv2.imshow('vertical',flipped_vertical)
    cv2.imshow('both', flipped_both)
    ##cv2.imwrite('/Users/karansingh/ocr-learning/Introduction/fruit_flipped.png', flipped_img) ## save flipped image
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
